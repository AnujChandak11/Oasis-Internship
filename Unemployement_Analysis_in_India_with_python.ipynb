**Exploratory Data Analysis**

```python
# Analysis Graph 1
# correlation in data
plt.style.use("seaborn-whitegrid")
plt.figure(figsize=(12,12))
sns.heatmap(df2.corr(),annot=True)
plt.show()
```

```python
df2.corr()
```

```python
# Analysis Graph 2
plt.figure(figsize=(30,8))
sns.countplot(x='Region.1',data=df2)
plt.show()
```

```python
# Analysis Graph 3
# Region
colour=sns.color_palette('pastel')
labels=df2['Region'].dropna().unique()
plt.figure(figsize=(18,10))
plt.subplot(1,2,1)
plt.title("Region_Percentage")
plt.pie(df2['Region'].value_counts(),labels=labels,autopct='%.2f%%')
plt.show()
```
